```python
L = []

for i in range(len(data['중식메뉴'].values)):
    a = re.findall('([/\w]+)\s', data['중식메뉴'].values[i])
    for j in range(len(a)):
        L.append(a[j])

S = pd.Series(L)
S
```

```python
food = pd.read_csv('food1.csv', header=2, index_col='NO', encoding='utf-8')
food = food.iloc[:, :23]
food.drop(['SAMPLE_ID', '식품코드', '연도', '채취시기', '총내용량(g)', '총내용량(mL)', '에너지(kj)', '수분(g)', '수분(%)', '지질-가식부 100g당', '지역 / 제조사'], axis=1, inplace=True)
food.drop(food[food['DB군']=='농축산물'].index, inplace=True)
food.drop(food[food['DB군']=='수산물'].index, inplace=True)
food = food.drop_duplicates(['식품명'], keep='first')
food.head()
```

```python
def menuLabeling(men):
    if len(food.loc[food['식품명'].str.contains(men), '식품대분류'].values) > 0:
        return food.loc[food['식품명'].str.contains(men), '식품대분류'].values.max()
    else:
        return np.nan
    
def menuLabeling2(men):
    if len(food.loc[food['식품명'].str.contains(men), '식품상세분류'].values) > 0:
        return food.loc[food['식품명'].str.contains(men), '식품상세분류'].values.max()
    else:
        return np.nan
```

```python
menu_label_df = pd.DataFrame({'menu': [men for men in S.unique()],
                              'label1': [menuLabeling(men) for men in S.unique()],
                              'label2': [menuLabeling2(men) for men in S.unique()]})
menu_label_df
```
